```python
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD

# Define the Bayesian network structure for Exercise 1
model_1 = BayesianModel([
    ('P', 'C'),  # Precio influences Compra
    ('G', 'C'),  # Género del Juego influences Compra
    ('CP', 'C'),  # Críticas Positivas influences Compra
    ('Pl', 'C'),  # Plataforma influences Compra
    ('A', 'C')    # Amigos que lo Juegan influences Compra
])

# Define the Bayesian network structure for Exercise 2
model_2 = BayesianModel([
    ('NE', 'EM'),  # Nivel de Experiencia influences Éxito de la Misión
    ('EQ', 'EM'),  # Equipamiento influences Éxito de la Misión
    ('H', 'EM'),   # Habilidades influences Éxito de la Misión
    ('S', 'EM'),   # Salud influences Éxito de la Misión
    ('EMe', 'EM')  # Estado Mental influences Éxito de la Misión
])

# Generate a simple plot for model_1
import networkx as nx
import matplotlib.pyplot as plt

# Exercise 1 Network Plot
plt.figure(figsize=(10, 6))
nx.draw(model_1, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold")
plt.title("Bayesian Network for Exercise 1")
plt.show()

# Exercise 2 Network Plot
plt.figure(figsize=(10, 6))
nx.draw(model_2, with_labels=True, node_size=2000, node_color="lightgreen", font_size=10, font_weight="bold")
plt.title("Bayesian Network for Exercise 2")
plt.show()
```
