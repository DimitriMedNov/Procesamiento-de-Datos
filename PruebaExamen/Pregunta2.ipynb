```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('pinturas.csv')

expected_columns = ['Comisiones', 'Llamadas', 'Millas']
if not all(column in df.columns for column in expected_columns):
    raise ValueError(f"El CSV debe tener las siguientes columnas: {expected_columns}")

sns.pairplot(df)
plt.show()

correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
from sklearn.linear_model import LinearRegression

data = {
    'Comisiones': [38, 44, 29, 38, 37, 14, 34, 25, 27, 25, 43, 34],
    'Llamadas': [146, 144, 147, 144, 149, 131, 144, 132, 132, 127, 154, 147],
    'Millas': [3290, 3103, 2122, 2791, 3209, 2287, 2848, 2690, 2933, 2671, 2988, 2829]
}
df = pd.DataFrame(data)

X = df[['Llamadas', 'Millas']] 
y = df['Comisiones']

model = LinearRegression()

model.fit(X, y)

coefficients = model.coef_
intercept = model.intercept_

print(f'Coeficientes: {coefficients}')
print(f'Interseccin: {intercept}')

r_squared = model.score(X, y)
print(f'Puntaje R^2: {r_squared}')
```
