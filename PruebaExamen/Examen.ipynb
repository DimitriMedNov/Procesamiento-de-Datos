```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('pregunta1.csv')

df.columns = ['Ingreso', 'Habitantes']

plt.figure(figsize=(10, 6))
sns.regplot(x='Habitantes', y='Ingreso', data=df)
plt.title('Grfica de correlacin lineal simple')
plt.xlabel('Nmero de habitaciones ocupadas')
plt.ylabel('Ingreso del restaurante')
plt.show()

correlation_coef = df['Habitantes'].corr(df['Ingreso'])
print(f"El coeficiente de correlacin de Pearson es: {correlation_coef}")


```
