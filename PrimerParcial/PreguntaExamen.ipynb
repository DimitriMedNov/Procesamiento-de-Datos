```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('distribuidor.csv', encoding='ISO-8859-1')
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()
cols = ['Nmero de tiendas', 'Nmero de automviles', 'Ingreso personal', 'Antigedad', 'Nmero de supervisores']
for i, col in enumerate(cols):
    axs[i].scatter(data[col], data['Ventas anuales'])
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Ventas anuales')
plt.tight_layout()
plt.show()
plt.scatter(data['Ingreso personal'], data['Ventas anuales'])
plt.xlabel('Ingreso personal (millones de dlares)')
plt.ylabel('Ventas anuales (millones de dlares)')
plt.show()
X = data['Ingreso personal'].values.reshape(-1,1)
Y = data['Ventas anuales'].values
model = LinearRegression()
model.fit(X, Y)
print("Coeficiente de regresin ('Ingreso personal'):", model.coef_[0])
print("Trmino independiente:", model.intercept_)
print("Varianza (score R^2):", model.score(X, Y))
dataX = data[['Nmero de tiendas', 'Nmero de automviles', 'Ingreso personal', 'Antigedad']]
model_multiple = LinearRegression()
model_multiple.fit(dataX, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Nmero de automviles'], data['Nmero de tiendas'], data['Antigedad'], c='r', marker='o')
ax.set_xlabel('Nmero de automviles')
ax.set_ylabel('Nmero de tiendas')
ax.set_zlabel('Antigedad')
plt.show()
print("Coeficientes de regresin:", model_multiple.coef_)
print("Varianza (score R^2) del modelo mltiple:", model_multiple.score(dataX, Y))
feature_names = ['Nmero de tiendas', 'Nmero de automviles', 'Ingreso personal', 'Antigedad']
X_new = pd.DataFrame([[2000, 11, 50, 4.5]], columns=feature_names)
predicted_value = model_multiple.predict(X_new)[0]
print("Prediccin para ingreso personal de 50, 11 automviles, 2000 tiendas y 4.5 de antigedad:", predicted_value)
```
