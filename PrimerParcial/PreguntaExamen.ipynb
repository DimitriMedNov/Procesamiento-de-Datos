```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('distribuidor.csv', encoding='ISO-8859-1')
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()
cols = ['Número de tiendas', 'Número de automóviles', 'Ingreso personal', 'Antigüedad', 'Número de supervisores']
for i, col in enumerate(cols):
    axs[i].scatter(data[col], data['Ventas anuales'])
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Ventas anuales')
plt.tight_layout()
plt.show()
plt.scatter(data['Ingreso personal'], data['Ventas anuales'])
plt.xlabel('Ingreso personal (millones de dólares)')
plt.ylabel('Ventas anuales (millones de dólares)')
plt.show()
X = data['Ingreso personal'].values.reshape(-1,1)
Y = data['Ventas anuales'].values
model = LinearRegression()
model.fit(X, Y)
print("Coeficiente de regresión ('Ingreso personal'):", model.coef_[0])
print("Término independiente:", model.intercept_)
print("Varianza (score R^2):", model.score(X, Y))
dataX = data[['Número de tiendas', 'Número de automóviles', 'Ingreso personal', 'Antigüedad']]
model_multiple = LinearRegression()
model_multiple.fit(dataX, Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Número de automóviles'], data['Número de tiendas'], data['Antigüedad'], c='r', marker='o')
ax.set_xlabel('Número de automóviles')
ax.set_ylabel('Número de tiendas')
ax.set_zlabel('Antigüedad')
plt.show()
print("Coeficientes de regresión:", model_multiple.coef_)
print("Varianza (score R^2) del modelo múltiple:", model_multiple.score(dataX, Y))
feature_names = ['Número de tiendas', 'Número de automóviles', 'Ingreso personal', 'Antigüedad']
X_new = pd.DataFrame([[2000, 11, 50, 4.5]], columns=feature_names)
predicted_value = model_multiple.predict(X_new)[0]
print("Predicción para ingreso personal de 50, 11 automóviles, 2000 tiendas y 4.5 de antigüedad:", predicted_value)
```
